```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([151, 174, 138, 186, 128, 136, 179, 163, 152, 131])
y = np.array([63, 81, 56, 91, 47, 57, 76, 72, 62, 48])
```

```python
x_mean = np.mean(x)
y_mean = np.mean(y)
print(x_mean,y_mean)
```

```python
beta1 = sum((x - x_mean) * (y - y_mean)) / sum((x - x_mean) ** 2)
print(beta1)
```

```python
beta0 = y_mean - beta1 * x_mean
print(beta0)
```

```python
x_pred = np.array([1, 2, 3])
y_pred = beta0 + beta1 * x_pred
print(x_pred,y_pred)
     
```

```python
plt.scatter(x, y)
plt.plot(x, beta0 + beta1*x, color='red')
plt.show()
```
